# With this script, one can extract the word list with translation from html of the duo website

You need to scroll down entirely, save the HTML code as '__filename__.txt' and update the filename and path
Remember in Windows to update / to //
To see the HTML Code, press Q (Firefox)

```python
from bs4 import BeautifulSoup
import html
from datetime import datetime

filename = 'C:\\Users\\janhe\\OneDrive\\Dokumente\\Python Scripts\\Duolingo\\Testfile.txt'  # Update with your file name

# Get the current date 
    # this is so one can update the list with new words without deleting any files from other scraping rounds
current_date = str(datetime.now().date())

def extract_vocabulary(filename):
    danish_vocabulary = []
    english_vocabulary = []
    
    with open(filename, 'r', encoding='utf-8') as file:
        text = file.read()
        soup = BeautifulSoup(text, 'html.parser')
        
        word_pairs = soup.find_all('div')
        
        for word_pair in word_pairs:
            danish_word = word_pair.find('h3')
            if danish_word:
                danish_vocabulary.append(html.unescape(danish_word.text.strip()))
            
            english_word = word_pair.find('p')
            if english_word:
                english_vocabulary.append(html.unescape(english_word.text.strip()).replace(',', '/'))
    
    return (danish_vocabulary[::3]), (english_vocabulary[::3]) # for some reason every word is in there 3 times


danish_vocabulary, english_vocabulary = extract_vocabulary(filename)

for i in range(len(danish_vocabulary)):
    print(danish_vocabulary[i], '\t', english_vocabulary[i])
```

# Merge and save the Data
File 'merged_vocabulary.txt'


```python
def merge_vocabulary(danish_vocabulary, english_vocabulary, output_file):
    with open(output_file, 'w', encoding='utf-8') as file:
        for danish_word, english_word in zip(danish_vocabulary, english_vocabulary):
            line = f"{danish_word}\t{english_word}\n"
            file.write(line)

# Example usage:
output_file = current_date + '_merged_vocabulary.txt'

merge_vocabulary(danish_vocabulary, english_vocabulary, output_file)
```

# Save Danish Pronounciation to the Anki Media Folder
Depending on your system. Find the folder with '%APPDATA%\Anki2' on Windows
This may take a few minutes


```python
from gtts import gTTS
import os

# Function to generate pronunciation audio and save it
def generate_and_save_pronunciation(danish_word, english_translation, output_folder):
    tts = gTTS(danish_word, lang='da')
    filename = 'danish_'+f"{danish_word}.mp3"
    filepath = os.path.join(output_folder, filename)
    tts.save(filepath)
    print(f"Saved audio for '{danish_word}' to '{filepath}'")

# Path to the merged vocabulary file
vocabulary_file = current_date + '_merged_vocabulary.txt'

# Folder to save the audio files
output_folder = 'C:\\Users\\janhe\\AppData\\Roaming\\Anki2\Benutzer 1\\collection.media'

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Read the merged vocabulary file
with open(vocabulary_file, "r", encoding="utf-8") as file:
    for line in file:
        danish_word, english_translation = line.strip().split("\t")
        generate_and_save_pronunciation(danish_word, english_translation, output_folder)
```
